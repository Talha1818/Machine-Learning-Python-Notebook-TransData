```python
df1 = df['label'].value_counts().to_frame().reset_index().rename(columns={'index':'label','label':'count'})
one_category = df1[df1['count']==1]
one_category_list= one_category['label'].tolist()
print(one_category_list)
```

```python
s = df[['label', 'probability']].values.tolist()

THRESHOLD = 0.8
for i in range(len(s)-1):
    if s[i][0] in one_category_list:
        pass
#         print(f"\nCategory 1 {s[i][0]} Prob {s[i][1]}\n")
    elif s[i][0]==s[i+1][0]:
        pass
#         print(f"Previous {s[i][0]}| Next {s[i+1][0]}")
    else:
        if (s[i][1]<THRESHOLD or s[i+1][1]<THRESHOLD) and not (s[i][0] in one_category_list or s[i+1][0] in one_category_list):
            prev_ = s[i][0]
            prev1 = s[i-1][0]
            next_ = s[i+1][0]
            next1 = s[i+2][0]
            next2 = s[i+3][0]
            print(f"Index:{i}  Previous1 {prev1} | Previous {prev_} Prob {s[i][1]} \t| Next {next_} Prob {s[i+1][1]} | Next1 {next1} | Next2 {next2}")
            if (prev_==next1 and prev_==next2) and (prev_==prev1):
                print("="*100)
                print(f"\t\tWe need to set {prev1} at INDEX {i+1}")
                print("="*100)              
                df.loc[i+1,'label'] = prev1
#                 df.loc[i,'probability'] = 0.7
```

```python
df = pd.read_csv('/content/test (1).csv')
from tabulate import tabulate
print(tabulate(df, headers = 'keys', tablefmt = 'psql'))
```

```python
df.loc[30, 'label'] = 34
df.loc[30, 'probability'] = 0.90

 
```

```python
#copying the dataframe to perform experiment
test_df = df.copy()


threshold = 0.97 #Setting the threshold value. Above than 96% seems to be good score

#List of documents that appear as a single page document in the package
single_list= [27, 29, 36, 38, 39, 44, 45, 614, 63, 71, 74, 
             76, 79, 80, 82, 84, 100, 103, 109, 110, 113, 117, 
             138, 197, 203, 75, 170, 102, 130, 160, 168, 183, 224, 228, 257]

#Initializing the variables
accurate = 0
total = 0


#getting the current accuracy of the package.

for i in range(test_df['label'].count()):
  if (test_df['label'][i] == test_df['actual'][i]):
    accurate +=1
  total +=1

print("score is: "+ str(round(accurate/total, 2)*100)+ "%")
```

```python
#Performing the actual algorithm. 


'''
We have restricted ourselves with a fact that 
eliminating false positives is a hard job to do.
Therefore, we are handling cases in which multipaged documents appears to be 
wrong document. This algorithm flaws only at one point i.e.
when we have single page document perdicted with higher accuracy above than threshold
and assigned label is also a single paged document but both are different.

Surprisingly, we are getting 1-2% accuracy improvements with postprocessing.

"actual" means what we actually got after passing from prediction
"probability" means confidence score assigned
"label" means label assigned to the document page
'''


for i in range(1, test_df['label'].count()-1):
  
  if (test_df['probability'][i] < threshold):
    if (int(test_df['label'][i]) not in single_list):
      if ((test_df['label'][i] != test_df['label'][i+1]) and (test_df['label'][i] != test_df['label'][i-1])):
        test_df['label'][i] = test_df['label'][i-1] 

        test_df['probability'][i] = 2.00 # This is redundant, unncessary. please ignore it while writing in the code
        


accurate = 0
total = 0

#Doing the same thing again

for i in range(test_df['label'].count()):
  if (test_df['label'][i] == test_df['actual'][i]):
    accurate +=1
  else:
      print(i, test_df['label'][i], test_df['actual'][i], test_df['probability'][i])
  total +=1

print("score is: "+ str(round(accurate/total, 2)*100)+ "%")

print(total-accurate)
```
